# How to embedd font for plotnine 

1. sh를 통해서 폰트를 설치하고, 패키지를 설치한다. 
2. 필요한 패키지를 로드하고 폰트 리프레시에 필요한 라이브러리 로드 후, 커널을 재로딩한다. 
3. 실행한다.

# Testing Font 

```python
from plotnine.data import mpg
from plotnine import ggplot, aes, facet_grid, labs, geom_point, theme

(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="엔진크기",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme(text=element_text(size=15, family='Nanum Brush Script'))
)
```

```python
from plotnine.data import mpg
from plotnine import ggplot, aes, facet_grid, labs, geom_point, theme_dark, theme_xkcd 

(
    ggplot(mpg)
    + facet_grid(facets="year~class")
    + aes(x="displ", y="hwy")
    + labs(
        x="Engine Size",
        y="Miles per Gallon",
        title="Miles per Gallon for Each Year and Vehicle Class",
    )
    + geom_point()
    + theme_xkcd()
)
```
